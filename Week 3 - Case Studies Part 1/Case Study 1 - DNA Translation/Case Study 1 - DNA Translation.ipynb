```python
def translate(seq):
    """Translates a string of nucleotides into a string of amino acids."""
    protein = ""
    if len(seq) % 3 == 0:
        for i in range(0, len(seq), 3):
            codon = seq[i:i+3]
            protein = protein + table[codon]
    return protein
```

```python
def read_seq(inputfile):
    """Reads and returns an input sequence with special characters removed."""
    with open(inputfile, "r") as f:
        seq = f.read()
    seq = seq.replace("\r", "")
    seq = seq.replace("\n", "")
    return seq
```
